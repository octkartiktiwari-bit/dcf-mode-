```python
print("welcome to my DCF model")
# Financial Inputs for Free Cash Flow Model

print("Enter company financial details:")

# Core operating inputs
ebit = float(input("Enter EBIT: "))
tax_rate = float(input("Enter tax rate (as %, e.g. 25 for 25%): "))

# Capital expenditure
capex = float(input("Enter CapEx (Purchase of PPE): "))

# Depriciation and amortisation
D_A= float(
    input(
        "Enter Depreciation and Amortization (In numbers): "
    )
 )
# Working capital inputs
current_assets_present = float(
    input("Enter Current Assets (Present Year): ")
)
current_assets_previous = float(
    input("Enter Current Assets (Previous Year): ")
)

current_liabilities_present = float(
    input("Enter Current Liabilities (Present Year): ")
)
current_liabilities_previous = float(
    input("Enter Current Liabilities (Previous Year): ")
)


print("\nInputs captured successfully.")
```

```python
#Present year Free Cash Flow

NOPAT=ebit*(1-(tax_rate/100))
Change_WC=(current_assets_present-current_liabilities_present)-(current_assets_previous-current_liabilities_previous)

Free_CF=NOPAT-capex+D_A-Change_WC
print(f"Present Year Free Cash Flow: {Free_CF}")
```

```python
#futute cash flow
# Efficiency metric
reinvested_value=((capex-D_A+Change_WC)/NOPAT)*100
print( f"Reinvested Value rate : {reinvested_value}")

print("For reference ->")
print("Company Type                     ROCE to Use")
print("------------------------------------------------")
print("PSU / mature industrial          18–22%")
print("Strong private industrial        20–25%")
print("Asset-light services             25–30%")
normalised_roce = float(
    input("Enter Normalised ROCE : ")
)
sustainable_GR=(reinvested_value*normalised_roce)/100
print(f"Sustainable Growth Rate : {sustainable_GR}")
print(f"Free Cash flow value on 1st year: {Free_CF}")
```

```python
#future cash flow
#free_cf1
#growthrate
g = sustainable_GR/100
# Map t = 0..5 to calendar years
year = 1
years = range(year, year + 6)

# Store results
fcf_values = {}
fcf_list=[]
for t, year in enumerate(years):
    fcf = Free_CF * (1 + g) ** t
    fcf_values[year] = fcf
    fcf_list.append(round(fcf, 2))

# Print results
for year, value in fcf_values.items():
    print(f"Year {year}: FCF = {value:.2f}")
```

```python
print("Let's find the WACC")

# Equity & Debt
equity_MC = 306499        # Market Capitalization
debt_BORROW = 61          # Borrowings
interest = 12             # Interest from P&L
taxes = 25                # Tax rate (%)

# CAPM Inputs
rf = 6.6                  # Risk-free rate (%)
rm = 12.6                 # Market return (%)
B = 0.7                   # Beta

# Cost of Debt
cost_of_debt = (interest / debt_BORROW) * 100
print(f"Cost of debt: {cost_of_debt}")

# Cost of Equity (CAPM)
cost_equity = rf + (B * (rm - rf))
print(f"Cost of equity: {cost_equity}")

# Capital Structure
EDE = equity_MC / (equity_MC + debt_BORROW)
DED = debt_BORROW / (equity_MC + debt_BORROW)

print(f"EDE: {EDE}")
print(f"DED: {DED}")

# WACC
WACC = (EDE * cost_equity + DED * cost_of_debt) * (1 - (taxes / 100))

print("WACC---------------->")
print(f"WACC: {WACC}")
```
